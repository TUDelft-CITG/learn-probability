## 1. Theory

Joint distributions describe the probability of multiple events occuring simultaneously. The joint probability distribution for two continuous variables is given by:

$f_{X Y}(x, y) d x d y=P(x<X \leq x+d x \bigcap y<Y \leq y+d y)$

The following rules apply to the joint PDF:

$0 \leq f_{X Y}(x, y)$, \
$\int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} f(x, y) d x d y=1 $, \
$\int_{-\infty}^{+\infty} f(x, y) d x =f(y) $, and $\int_{-\infty}^{+\infty} f(x, y) d y =f(x) $

According to the last condition (consistency rule), the marginal PDF of a variable is obtained if the joint PDF is integrated over the other variable.

The joint CDF of the two variables is given by:

$F_{X Y}(x, y) = \int_{-\infty}^{y} \int_{-\infty}^{x} f(x, y) d x d y = P(x<X \bigcap y<Y)$
